```python
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline
from rubin_sim.scheduler.model_observatory import ModelObservatory
from astropy.time import Time
```

```python
mo = ModelObservatory()
```

```python
conditions = mo.return_conditions()
```

```python
dir(conditions)
```

```python
hp.mollview(conditions.slewtime, unit='seconds', title='Slewtime', min=5, max=156)
plt.savefig('cond_slewtime.pdf')
```

```python
hp.mollview(conditions.skybrightness['r'], unit='$r$ mag arcsec$^{-2}$', title='Sky Brightness',
           max=21.3, min=19.8)
plt.savefig('cond_skybright.pdf')
```

```python
hp.mollview(conditions.airmass, unit='(unitless)', title='Airmass', max=3, min=1)
plt.savefig('cond_airmass.pdf')
```

```python
hp.mollview(conditions.fwhm_eff['r'], unit='arcsec', title='Effective Seeing, $r$', max=2, min=0.73)
```

```python
hp.mollview(np.degrees(conditions.solar_elongation), unit='degrees', title='Solar Elongation', max=180, min=0)
```
